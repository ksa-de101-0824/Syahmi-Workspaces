```python
"""
String method
1. capitalize() --> Convert the first character of a String to uppercase
2. lower() ---> Convert all characters of a string to lowercase
3. upper() ---> Convert all characters of a string to uppercase
4. title() ---> Convert the first character of each word to uppercase
5. split() ---> Splits a string into a list of substrings
6. strip() ---> Removes leading and trailing whitespaces
7. replace() ---> Replaces a substring with another substring
"""
#capitalize()
name = "abdul"
print("capitalize()")
print("Input : name = abdul")
print("Output:")
print(name.capitalize())
print("---------------")
#lower()
name = "SYAHMI"
print("lower()")
print("input : name - SYAHMI")
print("Output:")
print(name.lower())
print("---------------")
#upper()
name ="syahmi"
print("upper()")
print("Input : name = syahmi")
print("Output:")
print(name.upper())
print("---------------")
#title()
some_title = "malaysia won the first bronze medal in olympics 2024"
print("title()")
print("input : some_title = malaysia won the first bronze medal in olympics 2024 ")
print("Output:")
print(some_title.title())
print("---------------")
#split()
text="nama saya Syahmi"
print("split()")
print("Input : text = nama saya Syahmi")
print("Output:")
print(text.split())
print("---------------")
#strip()
text="      Malaysia is the best country to live in        "
print("strip()")
print("Output:")
print(text.strip())
print("---------------")
#replace()
text=("Malaysia won Bronze")
print("replace()")
print("Output:")
print(text.replace("Bronze" , "Silver"))
```
